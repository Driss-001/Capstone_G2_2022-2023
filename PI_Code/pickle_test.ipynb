```python
import pickle as pkl
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import os

current =  os.getcwd()
```

```python

X = np.linspace(0,5,500)
Y = np.cos(X)-np.sin(X)**2

XY = [X,Y]
with open(current+'\\data\\test.pkl', 'wb') as f:
    pkl.dump(XY,f)
plt.plot(X,Y)    
```

```python
with open(current+'\\data\\test.pkl', 'rb') as f:
    test_XY = pkl.load(f)

plt.plot(test_XY[0],test_XY[1])
```
